```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,'b')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('title')
```

###  Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])
```

```python
ax1.plot(x,y,'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(y,x,'g')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
fig
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])
```

```python
ax.plot(x,z,'b')
ax.set_xlabel('X')
ax.set_ylabel('Z')
ax2.plot(x,y,'b')
ax2.set_xlabel('X')
ax2.set_label('Y')
ax2.set_title('zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
fig
```
